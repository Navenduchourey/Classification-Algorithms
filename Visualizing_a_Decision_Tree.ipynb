**Hello World**
*   Claasification of Apples and oranges using small dataset.
*   Scikit learn DecisiionTreeClassifier is used to make predictions

```
from sklearn import tree

X = [ [140, 1], [146, 1], [158, 0], [160 , 0], [148, 1], [164, 0] ]
y = [ "Appple", "Apple", "Orange", "Orange", "Apple", "Apple" ]

clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y) 
predict = clf.predict([[143, 1]])
print(predict)
```

**Visualizing a DesicionTree Classifier**
*   Understanding IRIS dataset
*   List item



```
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
import numpy as np

iris = load_iris()
print(iris.feature_names)
print(iris.data[0])
print(iris.target_names)
print(iris.target[0])
```

```
import numpy as np
test_idx = [0, 50, 100]

train_data = np.delete( iris.data, test_idx, axis = 0 )
train_target = np.delete(iris.target, test_idx)
test_data = iris.data[test_idx]
test_target = iris.target[test_idx]
```

```
clf = DecisionTreeClassifier()
clf = clf.fit(train_data, train_target)
print(test_target)
predict = clf.predict(test_data)
print(predict)
```

```
from sklearn.externals.six import StringIO
from IPython.display import Image
import pydotplus

dot_data = StringIO()

tree.export_graphviz(clf, 
                     out_file=dot_data, 
                     feature_names= iris.feature_names, 
                     class_names=iris.target_names,
                     filled=True, impurity= False, rounded= True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
# graph.write_pdf("Iris.pdf")
Image(graph.create_png())
```
